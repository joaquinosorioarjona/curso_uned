```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
db_url = 'postgresql://postgres:postgres@localhost:5432/datos_telefonia'
engine = create_engine(db_url)
```

```python
query = """
    SELECT *
    FROM viajes_distritos_15_enero
    WHERE  gau_origen = 'GAU Sevilla'
    AND gau_destino = 'GAU Sevilla'
    AND renta = '<10' 
    AND hora > 6 AND hora < 21
"""

#Si quisiesemos hacer filtrado de fechas, AND fecha >= '2023-01-15 00:00:00' AND fecha < '2023-02-04 00:00:00'
```

```python
conn = psycopg2.connect(db_url)
```

```python
# Ejecutar la consulta y guardar los resultados en un DataFrame de pandas
df_viajes = pd.read_sql_query(query, conn)
df_viajes
```

```python
# Cerrar la conexión
conn.close()
```

```python
df_viajes.to_sql('viajes_sevilla_15_enero', engine, if_exists='replace', index=False)
```
